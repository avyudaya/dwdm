# Lab 11: Agglomerative Algorithm

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
data1 = np.array([[0.07, 0.83], [0.85, 0.14], [0.66, 0.89], [0.49, 0.64], [0.80, 0.46]])
# data2 = np.array([[18], [22], [25], [27], [42], [43]])
```

```python
data1
```

```python
# For complete link agglomerative clustering
dendrogram = sch.dendrogram(sch.linkage(data1, method='complete', metric='euclidean'))
```

```python
# For single link agglomerative clustering
dendrogram = sch.dendrogram(sch.linkage(data1, method='single', metric='euclidean'))
```

```python
# For average link agglomerative clustering
dendrogram = sch.dendrogram(sch.linkage(data1, method='average', metric='euclidean'))
```
