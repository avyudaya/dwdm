# Lab 4: K-Means Demonstration

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

```python
x = np.array([2, 2, 5, 4, 8, 6, 5, 5, 6, 4])
y = np.array([4, 6, 6, 7, 3, 6, 2, 7, 3, 4])
centroids = np.array([[1,5], [4,1], [8,4]])
centroids_x = centroids[:, 0]
centroids_y = centroids[:, 1]
```

### Visualize the points with scatterplot

```python
centroids_x = centroids[:, 0]
centroids_y = centroids[:, 1]
 
plt.figure(figsize=(15, 9))
plt.scatter(x, y, marker='o', color='grey', s=70)
plt.scatter(centroids_x, centroids_y, marker='x', color='red', s=100)
 
# Showing the coordinates 
for xy in zip(x, y):
 plt.annotate(' %d, %d' % xy, xy=xy)
plt.show()
```

```python
# Combined 2d data
X = np.transpose((x, y))
```

```python
X
```

### Clustering with Kmeans

```python
kmeans = KMeans(n_clusters=3, random_state=3)
y= kmeans.fit_predict(X)
centroids = kmeans.cluster_centers_
```

```python
y
```

```python
centroids
```

### Scatter plots for kmeans

```python
plt.figure(figsize=(15,9))
colors = ['yellow', 'green', 'grey']
for i, color in enumerate(colors):
 plt.scatter(X[y==i, 0], X[y==i, 1], c=color, label=f'Cluster {i+1}', s=50)
 
# plt.scatter(X[y==0, 0], X[y==0, 1], color='yellow', label='Cluster 1', s=50)
# plt.scatter(X[y==1, 0], X[y==1, 1], color='green', label='Cluster 2', s=50)
# plt.scatter(X[y==2, 0], X[y==2, 1], color='grey', label='Cluster 3', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color = 'red',
 label ='Centroids',s=100,marker='x')
plt.legend(loc='center right')
plt.show()
```
